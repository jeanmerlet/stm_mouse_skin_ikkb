```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.decomposition
import sklearn.manifold
import umap
import scprep
import time
```

```python
np.random.seed(888)
random_state = 888

data_dir = "/Users/6j9/projects/mouse/data/"
mtx_tsv_path = '/Users/6j9/projects/mouse/data/combined_mtx.tsv'
```

```python
cell_types_path = '/Users/6j9/projects/mouse/data/top8_v2.tsv'
cell_types = pd.read_csv(cell_types_path, sep='\t', header=None, index_col=0)
fibro_idx = (cell_types == 'Fibroblasts').values
fibro_bcs = cell_types.loc[fibro_idx, :].index.values

matrix = pd.read_csv(mtx_tsv_path, sep='\t', header=0, index_col=0)
matrix = matrix.transpose()
matrix = scprep.filter.filter_rare_genes(matrix, min_cells=10)
raw_values = matrix
matrix = scprep.normalize.library_size_normalize(matrix)
matrix = scprep.transform.sqrt(matrix)

fibro_mtx = matrix.loc[fibro_bcs, :]

labels = ['WT'] * 6599 + ['cKO'] * 4485
fibro_labels = ['WT'] * 3609 + ['cKO'] * 1527
```

```python
start = time.time()
pca_operator = sklearn.decomposition.PCA(n_components=100, random_state=random_state)
tsne_operator = sklearn.manifold.TSNE(n_components=2, random_state=random_state)
#Y_tsne = tsne_operator.fit_transform(pca_operator.fit_transform(np.array(matrix)))
Y_tsne = tsne_operator.fit_transform(pca_operator.fit_transform(fibro_mtx.values))
end = time.time()
print("Embedded t-SNE in {:.2f} seconds.".format(end-start))
```

```python
start = time.time()
reducer = umap.UMAP()
#UMAP = reducer.fit_transform(matrix.values)
UMAP = reducer.fit_transform(fibro_mtx.values)
end = time.time()
print("Embedded UMAP in {:.2f} seconds.".format(end-start))
```

# Case - Control

```python
cmap = ['dodgerblue', 'lightcoral']

fig, ax = plt.subplots(figsize=(8, 8))
scprep.plot.scatter2d(Y_tsne, label_prefix="t-SNE", title="Fibroblast Case vs. Control Projection",
                      c=fibro_labels, ticks=False, cmap=cmap, ax=ax, s=20, fontsize=14)
plt.tight_layout()
path = f'/Users/6j9/projects/mouse/plots/revisions/for_kang/fibroblasts_wt-cko.png'
plt.savefig(path, dpi=300)
#path = f'/Users/6j9/projects/mouse/plots/revisions/for_kang/fibroblasts_wt-cko.pdf'
#plt.savefig(path)
plt.show()
```

# CCL11 Expression Per Cell Type

```python
celltypes_path = '/Users/6j9/projects/mouse/data/revisions/cyto_export/multiple_celltypes_from_cyto.tsv'
celltypes = pd.read_csv(celltypes_path, sep='\t', header=0)
celltypes.columns = ['experiment', 'eleven_cts', 'jean-kang1', 'gdt_cells', 'fibroblast_subtypes', 'name', 'selected', 'shared_name']
celltypes = celltypes.drop(['jean-kang1', 'selected', 'shared_name'], axis=1)
uniq_cts = np.unique(celltypes.loc[:, 'eleven_cts'])
```

```python
uniq_cts
```

Need fibroblast subtypes from fibroblast_subtypes column and gamma delta T cells from gdt_cells column

```python
celltypes['combined_cts'] = celltypes['eleven_cts'].values
for i, ct in enumerate(celltypes['gdt_cells'].values.astype(str)):
    if ct == 'gd T cels':
        celltypes.iloc[i, 5] = ct
        
for i, ct in enumerate(celltypes['fibroblast_subtypes'].values.astype(str)):
    if ct in ['Dermal fibroblasts', 'Hypodermal fibroblasts']:
        celltypes.iloc[i, 5] = ct
```

```python
np.unique(celltypes['gdt_cells'].values.astype(str))
```

```python
np.unique(celltypes['fibroblast_subtypes'].values.astype(str))
```

```python
celltypes
```

```python
np.unique(celltypes['combined_cts'])
```

```python
ccl11 = raw_values.loc[:, 'CCL11']
ccl11 = ccl11[celltypes['name']]
combined_cts = celltypes['combined_cts']
uniq_cell_types = np.unique(combined_cts)
```

```python
ccl11_pos_counts = np.zeros(len(uniq_cell_types), dtype=np.int32)
for i, cell_type in enumerate(uniq_cell_types):
    bcs_idx = combined_cts == cell_type
    ccl11_pos_counts[i] = np.count_nonzero(ccl11.loc[bcs_idx.values])
```

```python
arg_idx = np.argsort(ccl11_pos_counts)[::-1]
ccl11_pos_counts[arg_idx]
```

```python
x = np.arange(len(ccl11_pos_counts))
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(x, ccl11_pos_counts[arg_idx])
ax.set_xticks(x)
ax.set_xticklabels(uniq_cell_types[arg_idx], rotation=90, size=14)
ax.tick_params(axis='y', which='major', labelsize=14)
ax.set_xlabel('Cell Type', size=18)
ax.set_ylabel('Number of Cells', size=18)
ax.set_title('CCL11 Expressing Cells', size=22)

plt.tight_layout()
fig_path = '/Users/6j9/projects/mouse/plots/revisions/for_kang/ccl11_bar_plot.pdf'
plt.savefig(fig_path)
plt.show()
```
